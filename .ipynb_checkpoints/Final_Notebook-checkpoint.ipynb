# Project Definition

## Project Overview

The Project is about creating a dog breed classifier. The model accepts an image and classifies the breed of that dog. Following are some of the characteristics of the project:
- It accepts an image. If there is a human present, it gives the most resembling dog breed to that human otherwise tells the breed of the dog
- The project is a popular one in the field of Artificial Intelligence and Machine Learning especially Deep learning (Convolutional Neural Networks).
- We would be using two separate datasets for this task. One that contains the dog images and their corresponding dog breeds and other that contains human images.
- The project will be based on accepting RGB images and will not be applicable for Grayscale images containing just one band.

## Problem Statement

The problem statement of this project can simply be stated as,
- **An algorithm that will accept any user-supplied image as input. If a dog is detected in the image, it will provide an estimate of the dog's breed. If a human is detected, it will provide an estimate of the dog breed that is most resembling**. 

An example of the end product can be as shown in the below picture.

![Sample Dog Output](static/sample_dog_output.png)

## Metrics

Following metrics will be used as a basis of measuring the correctness of the algorithm:
- The algorithm achieves an accuracy of 80% on the test dataset of dog images
- The algorithm achieves an f1-score of 80% on the test dataset of dog images
- The algorithm is able to detect a face 95% of the time on the human dataset

The above metrics are applicable and justified for our problem as Accuracy and F1-Score are the most used and beneficial metrics for classification algorithms

# Analysis

## Data Loading

First, we will load the data paths related to dog and human images. For this purpose, we will be using scikit-learn's `load_files` function. We will separately load images for train, test and validation sets

```python
from sklearn.datasets import load_files
import numpy as np
from glob import glob
import os
from keras.utils import np_utils
import random
```

```python
def load_dataset(dataset_path):
    '''Returns image paths and labels from the given path'''
    data = load_files(dataset_path)
    files = np.array(data['filenames'])
    targets = np_utils.to_categorical(np.array(data['target']), 133)
    return files, targets


train_files, train_targets = load_dataset('Data/dogImages/train')
valid_files, valid_targets = load_dataset('Data/dogImages/valid')
test_files, test_targets = load_dataset('Data/dogImages/test')
human_files = np.array(glob('Data/humanimages/*/*'))
random.shuffle(human_files)
```

Now we will write a function to convert numerical labels into their respective dog breeds.

```python
def label_to_category_dict(path):
    '''Returns a dictionary that maps labels to categories'''
    categories = os.listdir('Data/dogImages/train/')
    label_to_cat = map(lambda x: (int(x.split('.')[0]) - 1, x.split('.')[1]), categories)
    label_to_cat = {label: category for label, category in label_to_cat}
    return label_to_cat


label_to_cat = label_to_category_dict(train_files)
```

## Data Exploration

First, let's look at the numerical statistics about the datasets

```python
print('Number of categories: {}'.format(len(label_to_cat)))
print('Number of total dog images: {}'.format(len(train_files) + len(valid_files) + len(test_files)))
print('Number of training images: {}'.format(len(train_files)))
print('Number of validation images: {}'.format(len(valid_files)))
print('Number of test images: {}'.format(len(test_files)))
print('Number of Human Images: {}'.format(len(human_files)))
```

We will then look at the various counts of the dog breeds among the train, validation and test sets. We will write a function `categories_count` that will take a path containing different categories and will return the count of images in each category

```python
def categories_count(path):
    '''Returns dictionary containing count of images in each category in the path'''
    categories = os.listdir('Data/dogImages/train/')
    categories_dict = {}
    for category in categories:
        cat = category.split('.')[1]
        categories_dict[cat] = len(os.listdir(os.path.join(path, category)))
    return categories_dict


train_categories_dict = categories_count('Data/dogImages/train/')
test_categories_dict = categories_count('Data/dogImages/test/')
valid_categories_dict = categories_count('Data/dogImages/valid/')
```

Let's combine the count of all the sets to calculate the division of whole dataset

```python
total_categories = {}
for train_cat, train_count in train_categories_dict.items():
    for test_cat, test_count in test_categories_dict.items():
        for valid_cat, valid_count in valid_categories_dict.items():
            if train_cat == test_cat == valid_cat:
                   total_categories[train_cat] = train_count + test_count + valid_count
                    
# sort the categories according to their counts
total_categories = {k: v for k, v in sorted(total_categories.items(), key=lambda item: item[1], reverse=True)}
total_categories
```

Let's analyze the human dataset and try to detect a human face in the image

We use OpenCV's implementation of [Haar feature-based cascade classifiers](http://docs.opencv.org/trunk/d7/d8b/tutorial_py_face_detection.html) to detect human faces in images.  OpenCV provides many pre-trained face detectors, stored as XML files on [github](https://github.com/opencv/opencv/tree/master/data/haarcascades).  We have downloaded one of these detectors and stored it in the `../Data/haarcascades` directory.


```python
import cv2
import matplotlib.pyplot as plt

# extract pre-trained face detector
face_cascade = cv2.CascadeClassifier('Data/haarcascades/haarcascade_frontalface_alt.xml')

# load color (BGR) image
img = cv2.imread(human_files[3])

# convert BGR image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# find faces in image
faces = face_cascade.detectMultiScale(gray)

# print number of faces detected in the image
print('Number of faces detected:', len(faces))

# get bounding box for each detected face
for (x,y,w,h) in faces:
    # add bounding box to color image
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    
# convert BGR image to RGB for plotting
cv_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# display the image, along with bounding box
plt.imshow(cv_rgb)
plt.show()
```

Let's now write a function that takes input an image path and returns `True` if a face is detected and `False` otherwise

```python
def face_detector(img_path):
    '''Returns true if human face present in image'''
    face_cascade = cv2.CascadeClassifier('Data/haarcascades/haarcascade_frontalface_alt.xml')
    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray)
    return len(faces) > 0
```

Let's test the face detector by calculating the percentage of human faces detected in a list of image paths

```python
def percentage_of_faces_deteced(files_path):
    '''Returns percentages of images that have a human face detected'''
    preds = [face_detector(path) for path in files_path]
    percentage_of_human = (sum(preds) / len(files_path)) * 100
    return percentage_of_human


perc_humans = percentage_of_faces_deteced(human_files[0:100])
perc_dogs = percentage_of_faces_deteced(train_files[0:100])


print('Percentage of first 100 human images having a detected human face: {}'.format(perc_humans))
print('Percentage of first 100 dog images having a detected human face: {}'.format(perc_dogs))
```

# Methodology

## Data Preprocessing

Let's first start with loading the images from the given paths

```python
from keras.preprocessing import image                  
from tqdm import tqdm

def path_to_tensor(img_path):
    '''Converts the image in the given path to a tensor'''
    # loads RGB image as PIL.Image.Image type
    img = image.load_img(img_path, target_size=(224, 224))
    # convert PIL.Image.Image type to 3D tensor with shape (224, 224, 3)
    x = image.img_to_array(img)
    # convert 3D tensor to 4D tensor with shape (1, 224, 224, 3) and return 4D tensor
    return np.expand_dims(x, axis=0)

def paths_to_tensor(img_paths):
    '''Converts all the images in the given paths to tensors'''
    list_of_tensors = [path_to_tensor(img_path) for img_path in tqdm(img_paths)]
    return np.vstack(list_of_tensors)
```

We load all the images even if they are truncated and convert them to `float32`

```python
from PIL import ImageFile                            
ImageFile.LOAD_TRUNCATED_IMAGES = True                 

# pre-process the data for Keras
train_tensors = paths_to_tensor(train_files).astype('float32')
valid_tensors = paths_to_tensor(valid_files).astype('float32')
test_tensors = paths_to_tensor(test_files).astype('float32')
```

Now we initialize an `ImageDataGenerator` that rescales the images and also applies Data Augmentation for better results

```python
import tensorflow as tf
```

```python
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255,
                                                horizontal_flip=True,
                                                vertical_flip=True,
                                                rotation_range=20)

valid_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255.)

test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255.)


train_generator = train_datagen.flow(train_tensors, train_targets, batch_size=32)
valid_generator = train_datagen.flow(valid_tensors, valid_targets, batch_size=32)
test_generator = train_datagen.flow(test_tensors, test_targets, batch_size=32)
```

## Implementation & Refinement

First, we start by trying a custom Convolutional Neural Network for the classification of dog breeds

```python

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(224, 224, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(256, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(2048, activation='softmax'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1024, activation='softmax'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(133, activation='softmax')
])


model.summary()
```

Then we compile the model with `categorical_crossentropy` as the loss function and `rmsprop` as the optimizer. Finally, we train the model for 5 epochs

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

checkpointer = tf.keras.callbacks.ModelCheckpoint(filepath='../saved_models/weights_best_custom.hdf5', 
                               verbose=1, save_best_only=True)

model.fit(train_generator, epochs=5, validation_data=valid_generator, callbacks=[checkpointer])
```

Let us now calculate the f1 score and accuracy of the trained model

```python
from sklearn.metrics import accuracy_score, f1_score
```

```python
test_preds = model.predict(test_tensors)

accuracy = accuracy_score(np.argmax(test_targets, axis=1), np.argmax(test_preds, axis=1)) * 100
f1 = f1_score(np.argmax(test_targets, axis=1), np.argmax(test_preds, axis=1), average='weighted') * 100

print('Accuracy using Custom CNN: {}'.format(round(accuracy, 3)))
print('F1 Score using Custom CNN: {}'.format(round(f1, 3)))
```

The above model doesn't seem to perform quite well. We will now make use of __Transfer Learning__. We will use a VGG16 model pretrained on ImageNet. We discard the last layer of the model and attach a Dense layer which will predict 133 (number of dog breeds) classes.

```python
VGG16_model = tf.keras.applications.vgg16.VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

model = tf.keras.models.Sequential()
model.add(VGG16_model)
model.add(tf.keras.layers.GlobalAveragePooling2D())

model.add(tf.keras.layers.Dense(133, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

checkpointer = tf.keras.callbacks.ModelCheckpoint(filepath='../saved_models/weights_best_vgg16.hdf5', 
                               verbose=1, save_best_only=True)

early_stopping = tf.keras.callbacks.EarlyStopping(patience=5, monitor='val_loss')

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(patience=2, monitor='val_loss')

model.fit(train_generator, epochs=1, validation_data=valid_generator, callbacks=[checkpointer, early_stopping, reduce_lr])
```

```python
test_preds = model.predict(test_tensors)

accuracy = accuracy_score(np.argmax(test_targets, axis=1), np.argmax(test_preds, axis=1)) * 100
f1 = f1_score(np.argmax(test_targets, axis=1), np.argmax(test_preds, axis=1), average='weighted') * 100

print('Accuracy using VGG 16: {}'.format(round(accuracy, 3)))
print('F1 Score using VGG 16: {}'.format(round(f1, 3)))
```

Before concluding the whole project, we will first write a function that will be able to take as input an image, if a human is present, then output the most resembling dog breed otherwise give the breed of the dog


```python
def extract_Resnet50(tensor):
    '''Returns the VGG16 features of the tensor'''
    return tf.keras.applications.resnet50.ResNet50(weights='imagenet', include_top=False).predict(tf.keras.applications.resnet50.preprocess_input(tensor))


def predict_breed(img_path):
    '''Predicts the breed of the given image'''
    # extract bottleneck features
    bottleneck_feature = extract_Resnet50(path_to_tensor(img_path))
    bottleneck_feature = tf.keras.models.Sequential([
                            tf.keras.layers.GlobalAveragePooling2D(input_shape=bottleneck_feature.shape[1:])
                        ]).predict(bottleneck_feature).reshape(1, 1, 1, 2048)
    # obtain predicted vector
    predicted_vector = Resnet50_model.predict(bottleneck_feature)
    # return dog breed that is predicted by the model
    return label_to_cat[np.argmax(predicted_vector)]
```

```python
def dog_breed_classifier(image_path):
    '''
       Returns the breed of the dog in the image if present.
       If a human is present, predicts the most resembling dog breed
    '''
    is_face_detected = face_detector(image_path)
    dog_breed = predict_breed(image_path)
    if(is_face_detected):
        output = 'Human Face Detected. The most resembling dog breed to this human is: {}'.format(dog_breed)
    else:
        output = 'The most resembling dog breed to this dog is: {}'.format(dog_breed)
    print(output)
    return dog_breed
```
